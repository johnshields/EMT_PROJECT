# Wind Energy Research

## How wind energy is created?
The sun heats the earth unevenly, and this creates thermal air currents. To achieve equal temperatures around the earth, these air pockets move the about as global wind. The energy that travels in the wind can be captured and converted to provide electricity.

#### Clean and sustainable energy
Wind energy provides a clean, sustainable solution to our energy problems. It can be used as an alternative to fossil fuels in generating electricity without the direct emission of greenhouse gases. And there will always be wind; it is inexhaustible and renewable.

## Wind turbine measuring wind speed
The amount of electricity a wind turbine can generate depends on the local wind speed. The wind speed itself depends
on several factors such as:

* Whether there are any obstructions such as trees and buildings nearby (which slow the wind-down and cause turbulence).

* The height above ground level: wind speeds increase with height so that the higher a turbine is, the more electricity it is likely to produce. 

* The location of a wind turbine is therefore crucial for maximizing its overall performance.

### Why do the turbines not spin at times?
The most common reason that turbines stop spinning is that the wind is not blowing fast enough.  Most wind turbines need a sustained wind speed of 9 MPH or higher to operate. Technicians will also stop turbines from performing routine maintenance or repairs.

# Training the Models

## The Power Production Data Set

```python
# necessary imports
# data frames
import pandas as pd
# plotting
import matplotlib.pyplot as plt
# plot style
plt.style.use("ggplot")
# plot size
plt.rcParams['figure.figsize'] = [14, 8]

# load dataset from csv file
powerproduction = pd.read_csv("https://tinyurl.com/y8s579fs")
```

### Sample of the Data Set

```python
# turn dataset into dataframe
ppDataSet = pd.DataFrame(powerproduction)

# output dataFrame sample
ppDataSet.sample()
```

### Data Set Graph

```python
# plot dataset
plt.title('Power Production Data Set')
plt.plot(ppDataSet, 'o', markersize=2)
plt.plot(ppDataSet['speed'], 'o', color="#fc7405", markersize=3, label='Wind Speed')
plt.plot(ppDataSet['power'], 'o', color="#1b55b3", markersize=3, label='Power')
plt.ylabel('Wind Turbine Power')
plt.xlabel('Wind Speed')
plt.legend()
```

The Data Set can be viewed [here](https://raw.githubusercontent.com/ianmcloughlin/2020A-machstat-project/master/dataset/powerproduction.csv). 

## Training Model Attempts

```python
# necessary imports
# neural networks
import tensorflow.keras as krs
from tensorflow.python.keras.layers import Dense
from tensorflow.python.keras.models import Model
# numerical arrays
import numpy as np
# make numpy values easier to read
np.set_printoptions(precision=3, suppress=True)
```

## Train Model: Attempt 1

Building a Sequential Model that has a Sigmoid and a Linear Layer.

Compile the Model with the Optimizer that implements the Adam algorithm. 

```python
# train model
model = krs.models.Sequential()
model.add(Dense(50, input_shape=(1,), activation='sigmoid', kernel_initializer="glorot_uniform", bias_initializer="glorot_uniform"))
model.add(Dense(1, activation='linear', kernel_initializer="glorot_uniform", bias_initializer="glorot_uniform"))
# compile model
model.compile(krs.optimizers.Adam(lr=0.001), loss='mean_squared_error', metrics=['accuracy'])
```

Fitting the Predicted Data with the Colums from the Data Set.

```python
# fit the data
m1history = model.fit(ppDataSet['speed'], ppDataSet['power'], epochs=500, batch_size=10)
```

### Model 1 Summary

```python
model.summary()
```

#### Head of Predicted Data

```python
# turn prediction 1 into data frame
dataPredict1 = pd.DataFrame(model.predict(ppDataSet['speed']), columns=['speed'])

# output head of prediction
dataPredict1.head()
```

Saving the Model to a .h5 file.

```python
# save prediction result to a .h5 file
model.save("data_files\mp.h5")
print("Saved model prediction to directory")
```

###### END OF ATTEMPT  1

## Train Model: Attempt 2

Loading the Data Set as NumPy arrays.

```python
# dataset as numpy arrays
x = ppDataSet.iloc[:,0].values
y = ppDataSet.iloc[:,1].values

xArray = np.array(x)
yArray = np.array(y)
```

Buidling the Model using 6 layers. Layer 1 is a Elu, Layer 2 and 3 are Tanh and 4, 5 and 6 are Relu Layers. Layer 6 uses Layer 5 as an output Layer.

```python
# build model using more layers
krsInput = krs.Input(shape=(1,))
layer1 = Dense(10, activation='elu')(krsInput)
layer2 = Dense(50, activation='tanh')(layer1)
layer3 = Dense(50, activation='tanh')(layer2)
layer4 = Dense(50, activation='relu')(layer3)
layer5 = Dense(50, activation='relu')(layer4)
layer6 = Dense(50, activation='relu')(layer5)
output = Dense(1)(layer6)
```

Compile the Model with the Input and Outputs from the Layers above.

This Model also uses the Adam Optimizer.

```python
# compile model
model2 = Model(inputs=krsInput, outputs=[output])
model2.compile(optimizer='adam', loss=['mean_squared_error'], metrics=['accuracy'])
```

Fitting the Data Arrays to the Model and then the Model is told to Predict Data from the Arrays.

```python
# fit data arrays
m2history = model2.fit(xArray, yArray, epochs=5000, batch_size=500)
model2.predict([1.001])
```

### Model 2 Summary

```python
model2.summary()
```

#### Head of Predicted Data

```python
# turn prediction 2 into data frame
dataPredict2 = pd.DataFrame(model2.predict(ppDataSet['speed']), columns=['speed'])

# output head of prediction
dataPredict2.head()
```

Saving the Model to a .h5 file.

```python
model2.save("data_files\mp2.h5")
print("Saved model prediction to directory")
```
