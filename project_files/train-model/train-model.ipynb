# Training the Models
***

```python
# necessary imports
# neural networks
import tensorflow.keras as krs
from tensorflow.python.keras.layers import Dense
from tensorflow.python.keras.models import Model
# numerical arrays
import numpy as np
# make numpy values easier to read
np.set_printoptions(precision=3, suppress=True)
# data frames
import pandas as pd
# plotting
import matplotlib.pyplot as plt

# plot style
plt.style.use("ggplot")
# plot size
plt.rcParams['figure.figsize'] = [14, 8]
```

```python
# load dataset from csv file
powerproduction = pd.read_csv("https://tinyurl.com/y8s579fs")
```

#### Sample of Original Data from the Data Set

```python
# turn dataset into dataframe
ppDataSet = pd.DataFrame(powerproduction)

# output dataFrame sample
ppDataSet.sample()
```

## Train Model: Attempt 1

```python
# train model
model = krs.models.Sequential()
model.add(Dense(50, input_shape=(1,), activation='sigmoid', kernel_initializer="glorot_uniform", bias_initializer="glorot_uniform"))
model.add(Dense(1, activation='linear', kernel_initializer="glorot_uniform", bias_initializer="glorot_uniform"))
# compile model
model.compile(krs.optimizers.Adam(lr=0.001), loss='mean_squared_error', metrics=['accuracy'])
```

```python
# fit the data
m1history = model.fit(ppDataSet['speed'], ppDataSet['power'], epochs=500, batch_size=10)
```

### Model 1 Summary

```python
model.summary()
```

#### Head of Predicted Data

```python
# turn prediction 1 into data frame
dataPredict1 = pd.DataFrame(model.predict(ppDataSet['speed']), columns=['speed'])

# output head of prediction
dataPredict1.head()
```

```python
# save prediction result to a .h5 file
model.save("data_files\mp.h5")
print("Saved model prediction to directory")
```

###### END OF ATTEMPT  1

## Train Model: Attempt 2

```python
# dataset as numpy arrays
x = ppDataSet.iloc[:,0].values
y = ppDataSet.iloc[:,1].values

xArray = np.array(x)
yArray = np.array(y)
```

```python
# build model using more layers
krsInput = krs.Input(shape=(1,))
layer1 = Dense(10, activation='relu')(krsInput)
layer2 = Dense(50, activation='relu')(layer1)
layer3 = Dense(50, activation='relu')(layer2)
layer4 = Dense(50, activation='relu')(layer3)
layer5 = Dense(50, activation='relu')(layer4)
layer6 = Dense(50, activation='relu')(layer5)
output = Dense(1)(layer6)
```

```python
# compile model
model2 = Model(inputs=krsInput, outputs=[output])
model2.compile(optimizer='adam', loss=['mean_squared_error'], metrics=['accuracy'])
```

```python
# fit data arrays
m2history = model2.fit(xArray, yArray, epochs=30000, batch_size=500)
model2.predict([1.001])
```

### Model 2 Summary

```python
model2.summary()
```

#### Mean of Predicted Data

```python
# turn prediction 2 into data frame
dataPredict2 = pd.DataFrame(model2.predict(ppDataSet['speed']), columns=['speed'])

# output mean of prediction
dataPredict2.mean()
```

```python
model2.save("data_files\mp2.h5")
print("Saved model prediction to directory")
```

###### END OF ATTEMPT  2

#### END OF TRAINING MODELS

## Comparing Model 1 and 2

### Model 1

Model 1 has an average accuracy of 0.017324 over 500 Epochs.

### Accuracy of Model 1 

```python
# plot accuracy of model 1
plt.plot(m1history.history['accuracy'])
plt.title('Model 1 Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train'], loc='upper right')
```

### Mean of Accuracy for Model 1

```python
# turn model 1 accuracy into data frame
model1Acc = pd.DataFrame(m1history.history['accuracy'], columns=['Model Accuracy ='])

# output mean of model accuracy 
model1Acc.mean()
```

### Losses of Model 1

The Model goes off track with the losses, which the minimum is 254.019913.

```python
# plot the losses of model 1
plt.plot(m1history.history['loss'])
plt.title('Model 1 Losses')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train'], loc='upper right')
```

### Minium Loss for Model 1

```python
# turn model losses into data frame
model1Loss = pd.DataFrame(m1history.history['loss'], columns=['Minium Loss ='])

# output minium loss from model
model1Loss.min()
```

### Original Data vs Model 1's Predicted Data

The Predicted Data from the Model seems to follow the Original Data when plotted. However it does not match the Original Data Accurately.

```python
# plot model 1
plt.title('Original Data and Predicted Data from Model 1')
plt.plot(ppDataSet['speed'], ppDataSet['power'], 'o', color="#636362", markersize=2, label='Original Data')
plt.plot(ppDataSet['speed'], model.predict(ppDataSet['speed']), '-', color="#91130a", markersize=2, label='Prediction')
plt.legend();
```
